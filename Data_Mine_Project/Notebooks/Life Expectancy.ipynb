# Life Expectancy Notebook

This notebook is designed to give us a look into how life expectancy in the United States has evolved over time so we can better guage the impacts of our project over time. Please note that we had to clean our dataset so we could differentiate between gender and race in the US.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

lifeData = pd.read_csv('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/CSV Files/LEAB.csv')
```

```python
lifeNulls = lifeData.isnull().sum().sum()  
print(f"Total null values in life expectancy Data: {lifeNulls}")
```

```python
null_columns = lifeData.isnull().sum()
print(null_columns[null_columns > 0])
```

```python
lifeData = lifeData.dropna()
```

```python
totalAveLife = lifeData[
    (lifeData['Race'] == 'All Races') &
    (lifeData['Sex'] == 'Both Sexes')] ['Average Life Expectancy (Years)']


totalFemaleLife = lifeData[
    (lifeData['Race'] == 'All Races') &
    (lifeData['Sex'] == 'Female')] ['Average Life Expectancy (Years)']

totalMaleLife = lifeData[
    (lifeData['Race'] == 'All Races') &
    (lifeData['Sex'] == 'Male')] ['Average Life Expectancy (Years)']


uniqueYears = np.arange(1900,2019)

plt.title('Average Life Expectancy Among All Adults (United States)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)


# All Life Expectancies
plt.plot(uniqueYears, totalAveLife, label='Total Average Life Exectancy', color='r')
plt.plot(uniqueYears, totalFemaleLife, label='Female Life Exectancy', color='g')
plt.plot(uniqueYears, totalMaleLife, label='Male Life Exectancy', color='b')



# Adding a grid for better readability
plt.grid(True)
    
# Show legend
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)

# Run only when you need a new copy of the image. 
#plt.savefig('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/Visuals/AllAdultLfeExpectancy.png')

plt.show()
```

```python
allBlack = lifeData[
    (lifeData['Race'] == 'Black') &
    (lifeData['Sex'] == 'Both Sexes')] ['Average Life Expectancy (Years)']


femaleBlack = lifeData[
    (lifeData['Race'] == 'Black') &
    (lifeData['Sex'] == 'Female')] ['Average Life Expectancy (Years)']

maleBlack = lifeData[
    (lifeData['Race'] == 'Black') &
    (lifeData['Sex'] == 'Male')] ['Average Life Expectancy (Years)']


uniqueYears = np.arange(1900,2018)

plt.title('Average Life Expectancy Among All Black Adults (United States)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)


# All Life Expectancies
plt.plot(uniqueYears, allBlack, label='Total Average Life Exectancy', color='mediumorchid')
plt.plot(uniqueYears, femaleBlack, label='Female Life Exectancy', color='mediumturquoise')
plt.plot(uniqueYears, maleBlack, label='Male Life Exectancy', color='y')



# Adding a grid for better readability
plt.grid(True)
    
# Show legend
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)

# Run only when you need a new copy of the image. 
#plt.savefig('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/Visuals/AllBlackAdultLfeExpectancy.png')

plt.show()
```

```python
allWhite = lifeData[
    (lifeData['Race'] == 'White') &
    (lifeData['Sex'] == 'Both Sexes')] ['Average Life Expectancy (Years)']


femaleWhite = lifeData[
    (lifeData['Race'] == 'White') &
    (lifeData['Sex'] == 'Female')] ['Average Life Expectancy (Years)']

maleWhite = lifeData[
    (lifeData['Race'] == 'White') &
    (lifeData['Sex'] == 'Male')] ['Average Life Expectancy (Years)']


uniqueYears = np.arange(1900,2018)

plt.title('Average Life Expectancy Among All White Adults (United States)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)


# All Life Expectancies
plt.plot(uniqueYears, allWhite, label='Total Average Life Exectancy', color='magenta')
plt.plot(uniqueYears, femaleWhite, label='Female Life Exectancy', color='darkorange')
plt.plot(uniqueYears, maleWhite, label='Male Life Exectancy', color='lawngreen')



# Adding a grid for better readability
plt.grid(True)
    
# Show legend
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)

# Run only when you need a new copy of the image. 
#plt.savefig('/home/jovyan/DataMiningProject/DataMineProject/Data_Mine_Project/Visuals/AllWhiteAdultLfeExpectancy.png')

plt.show()
```
